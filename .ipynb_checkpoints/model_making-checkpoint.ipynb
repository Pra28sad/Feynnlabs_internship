### Step 5: Data Scaling


```python
scaler = StandardScaler()
```

```python
scaler
```

```python
categorical_columns=['yummy',
 'convenient',
 'spicy',
 'fattening',
 'greasy',
 'fast',
 'cheap',
 'tasty',
 'expensive',
 'healthy',
 'disgusting']
X_scaled = scaler.fit_transform(df[categorical_columns])
```

```python
X_scaled
```

```python
df
```

### Step 6: Dimensionality Reduction using PCA (for visualization purposes)

```python

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c='red', edgecolor='k')
plt.title('PCA of McDonald\'s Data')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

### Step 7: Finding the Optimal Number of Clusters (Using Elbow Method)

```python

inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
```

```python

plt.figure(figsize=(8, 6))
plt.plot(K, inertia, 'bo-', markersize=8)
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

### Step 8: Applying K-Means Clustering (assuming optimal K=3 from the elbow method)


```python
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

### Step 9: Evaluation - Silhouette Score

```python
silhouette_avg = silhouette_score(X_scaled, df['Cluster'])
print(f'Silhouette Score for K={optimal_k}: {silhouette_avg}')
```

### Step 10: Visualizing Clusters in PCA space 

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df['Cluster'], cmap='viridis', edgecolor='k', s=100)
plt.title('Customer Segments (K-Means Clustering)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar()
plt.show()
```

### Step 11: Profiling the Segments

```python
cluster_profile = df.groupby('Cluster').mean()
print("\nCluster Profiles (Average for each feature):")
print(cluster_profile)
```

```python
# Visualize segment profiles
cluster_profile[categorical_columns].plot(kind='bar', figsize=(10, 6), stacked=True)
plt.title('Cluster Profiles for Binary Variables')
plt.ylabel('Proportion of "Yes" Responses')
plt.xlabel('Cluster')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

### Step 12: Further Evaluation - Analyze each segment's attributes

```python
# For example, let's compare Age and Visit Frequency across clusters
plt.figure(figsize=(8, 6))
sns.boxplot(x='Cluster', y='Age', data=df)
plt.title("Age Distribution by Cluster")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Cluster', hue='VisitFrequency', data=df)
plt.title("Visit Frequency by Cluster")
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import fcluster

# Step 13: Hierarchical Clustering with Dendrogram
# Compute the linkage matrix
linked = linkage(X_scaled, method='ward')

# Plot the dendrogram
plt.figure(figsize=(25,8))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title('Dendrogram of McDonald\'s Data')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distance')
plt.show()

# Step 14: Assign Clusters based on Dendrogram (cutting the tree)
# Set the number of clusters you want (based on the dendrogram's shape)
hierarchical_clusters = fcluster(linked, t=3, criterion='maxclust')  # `t` is the threshold or number of clusters

# Add the hierarchical clusters to the DataFrame
df['Hierarchical_Cluster'] = hierarchical_clusters

# Step 15: Visualize the clusters (optional)
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df['Hierarchical_Cluster'], cmap='plasma', edgecolor='k', s=100)
plt.title('Hierarchical Clustering of McDonald\'s Data (3 clusters)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar()
plt.show()

# Step 16: Analyze and Profile the Hierarchical Clusters
hierarchical_cluster_profile = df.groupby('Hierarchical_Cluster').mean()
print("\nHierarchical Cluster Profiles (Average for each feature):")
print(hierarchical_cluster_profile)

# Visualize the profiles of hierarchical clusters
hierarchical_cluster_profile[categorical_columns].plot(kind='bar', figsize=(10, 6), stacked=True)
plt.title('Hierarchical Cluster Profiles for Binary Variables')
plt.ylabel('Proportion of "Yes" Responses')
plt.xlabel('Hierarchical Cluster')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```
